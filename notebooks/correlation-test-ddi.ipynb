```python
all_candidates_train = load_pkl('cache/pkl/v1/candidates.train.pkl')
all_candidates_test = load_pkl('cache/pkl/v1/candidates.test.pkl')
```

```python
pair_list = list()
for c in all_candidates_train:
    pair_list.append((c['e1']['@text'], c['e2']['@text']))
```

```python
unique_pair_list = set(pair_list)
```

```python
from tqdm import tqdm
```

```python
labels_dict = dict()
for c in tqdm(unique_pair_list):
    labels_dict[c] = list()
    for i in range(len(pair_list)):
        if c == pair_list[i]:
            labels_dict[c].append(all_candidates_train[i]['label'])
```

```python
for i in all_candidates_train:
    if i['e1']['@text'] == 'phenytoin' and i['e2']['@text'] == 'contraceptives':
        print(i)
```

```python
diff_counter = 0
counter = 0
for k, v in labels_dict.items():
    if len(v) > 1 and len(list(set(v))) != 1:
        print(k, v)
        diff_counter += 1
        counter += 1
    elif len(v) > 1 and len(list(set(v))) == 1:
        counter += 1
```
